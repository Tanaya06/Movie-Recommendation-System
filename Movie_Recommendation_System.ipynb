```python
# Import Libraries

import numpy as np
import pandas as pd
import nltk
```

```python
# Read dataset

movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
# Merge 'movies' and 'credits' on 'title'
# Both datasets can also merged on 'id'
movies = movies.merge(credits, on='title')
```

####**DATA PREPROCESSING**

```python
movies.info()
```

```python
# Remove unneccessary columns

movies_df = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_df.head()
```

```python
# Find missing data

movies_df.isnull().sum()
```

```python
# Drop Null rows

movies_df.dropna(inplace = True)
```

```python
# Find duplicate values

movies_df.duplicated().sum()
```

```python
# Function to convert string obj of genre and keywords to list
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L 
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df.head()
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
# Function to get top 3 cast
def convert_cast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L 
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert_cast)
movies_df.head()
```

```python
# Function to get Director Name
def get_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L 
```

```python
movies_df['crew'] = movies_df['crew'].apply(get_director)
movies_df.head()
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
movies_df.head()
```

```python
#Replace spaces between column values

movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
# Add a new column to the dataset as 'tags'

movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df.head()
```

```python
# Make a new dataframe of only 'id', 'title' and 'tags'

new_df = movies_df[['id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
# Covert tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) 
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Function to reduce repeated words. Same words in different tense

def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  
  return " ".join(y)
```

```python
# Apply stem() function to new_df

new_df['tags'] = new_df['tags'].apply(stem)
new_df.head()
```

```python
# Convert tags into text vectors
# Remove the stop words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
text_vectors = cv.fit_transform(new_df['tags']).toarray()
text_vectors
```

```python
text_vectors.shape
```

```python
cv.get_feature_names()
```

```python
# Get similarities in movies by calculating cosine distance

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(text_vectors)
similarity_matrix
```

```python
similarity_matrix.shape
```

```python
sorted(list(enumerate(similarity_matrix[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
# Function to find the most similar movies to recommend

def recommend(movie):
  movie_idx = new_df[new_df['title'] == movie].index[0]
  distances = similarity_matrix[movie_idx]
  # Sort the movies in descending order to find similar movies
  # Use enumerate to keep track of movie index
  similar_movies = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

  # Print recommended movie indexes - Consider only TOP 5 similar movies
  for i in similar_movies:
        print(new_df.iloc[i[0]].title)
```
